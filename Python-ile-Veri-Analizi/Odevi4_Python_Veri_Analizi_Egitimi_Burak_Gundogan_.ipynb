### Veriyi özetleme

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px 
import plotly.graph_objects as go
```

1.  3. Ders sonunda hazırladığınız 2021 ve 2022 yıllarına ait BIST30 ve EUR/USD kur değerlerini içeren bileşke tablodan sadece ayın son işgününe ait değerleri içeren bir pivot tablo üretiniz.

```python
kurToplam=yf.download("EURTRY=X",start="2021-01-01",end="2022-12-31")

bist30Toplam=yf.download("XU030.IS",start="2021-01-01",end="2022-12-31")
```

```python
bist30Toplam=bist30Toplam["Adj Close"]
bist30Toplam=pd.DataFrame(bist30Toplam)
bist30Toplam.columns=["BIST30 Kapanış (TRY)"]
bist30Toplam
```

```python
kurToplam=kurToplam["Adj Close"]
kurToplam=pd.DataFrame(kurToplam)
kurToplam.columns=["EUR/TL Kapanış (TRY)"]
kurToplam
```

```python
mergeDf=pd.concat([kurToplam,bist30Toplam],axis=1)

mergeDf
```

```python
mergeDf['Date'] = pd.to_datetime(mergeDf.index)

# Sadece ayın son iş günlerine ait verileri seçme işlemi
sonIsGunu = mergeDf.groupby([mergeDf['Date'].dt.year, mergeDf['Date'].dt.month]).tail(1)

pivot=pd.pivot_table(sonIsGunu,values=["BIST30 Kapanış (TRY)","EUR/TL Kapanış (TRY)"],index=sonIsGunu.index)

pivot
```

2. Ay kırılımlı olarak ürettiğiniz tabloyu grafik olarak inceleyiniz.


```python
pivotPlt=px.line(pivot,color_discrete_map={"EUR/TL Kapanış (TRY)":"red","BIST30 Kapanış (TRY)":"blue"})
pivotPlt
```

3. Ders sonunda hazırladığınız 2021 ve 2022 yıllarına ait BIST30 ve EUR/USD kur değerlerini içeren bileşke tablodan sadece ayın son işgününe ait değerleri içeren bir tabloyu bir de resample yöntemi ile üretiniz.

```python
mergeDf=mergeDf.resample(rule="M").sum()
mergeDf
```

4. Resample yöntemi ile ürettiğiniz tabloyu grafik olarak inceleyiniz. 2. adımda elde ettiğiniz grafik ile karşılaştırarak yorumlayınız.

```python
mergeDfPlt=px.line(mergeDf,color_discrete_map={"EUR/TL Kapanış (TRY)":"red","BIST30 Kapanış (TRY)":"blue"})
mergeDfPlt
```
