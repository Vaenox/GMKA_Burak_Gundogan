### Bag of Words (BoW YÃ¶ntemiyle Kelimeleri VektÃ¶r Haline Getirme)

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer #bow modeli
import onIsleme
```

```python
df["Ä°ÅŸlenmiÅŸ Text"]=df["Text"].apply(onIsleme.preProcessing)
df["Ä°ÅŸlenmiÅŸ Text"]=df["Ä°ÅŸlenmiÅŸ Text"].apply(onIsleme.removeSpace)
```

```python
#boÅŸ liste kontrolÃ¼

df[df["Ä°ÅŸlenmiÅŸ Text"].str[0].isnull()]
```

```python
#boÅŸ satÄ±rlarÄ± silme

df_index=df[df["Ä°ÅŸlenmiÅŸ Text"].str[0].isnull()].index
df=df.drop(df_index)
df=df.reset_index()
del df["index"]

df[df["Ä°ÅŸlenmiÅŸ Text"].str[0].isnull()]
```
